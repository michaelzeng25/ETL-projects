```python
import requests
import urllib.request
from bs4 import BeautifulSoup as bs
```

# grab the contents from the website

```python
# set up a vsriable for the link where the data will be obtained.
url = 'https://historicalstockinfo.com/att-corp-stock-prices-table/'
r = requests.get(url)

# =200 if the hdmi codes are downloaded.
r.status_code
```

```python
# this will give all the information from the link
r.content
```

```python
# Organize the contents so they will appear like a block of hdmi code
soup = bs(r.content, 'html.parser')
soup
```

```python
# grabed a single table
find_table = soup.find_all('table')
print('number of tables:', len(find_table))
find_table
```

```python
# set the table to a variable
table = find_table[0]
```

```python
all_tr = table.find_all('tr')
all_tr
```

```python
# print the data in each row.
for row in all_tr:
    for cell in row.find_all('td'):
        print(cell.text, end=' ')
    print('\n')
    
```

# Writ the data into either csv or txt file

```python
# write the data into a csv or txt file sep by ,
with open('ATT Corp Stock Prices.csv','w') as r:
    for row in all_tr:
        for cell in row.find_all('td'):
            r.writelines(cell.text+",") # seperate each value by ',' so it can be read easily
        r.write('\n') # this will make sure that the reading ends at the end of each row in the table      
```

# Process the data using some powerful python libraries

```python
import pandas as pd
df = pd.read_csv('ATT Corp Stock Prices.txt', sep=',',index_col=False,names=['Year','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
df
```

```python
# The main contents of this table are the date and each values. 
# 1, the rows should be remove if its values in the year column is not number or date
# 2, convert the data are the right format. if the data are already in the right format, iterate throut it to confirm.
```

```python
# save the trouble rows' index to a array
ind = []
for i in range(0,38):
    if df['Year'].iloc[i].isdigit()==False:
        ind.append(i)
        
# drop the rows
df.drop(index=ind, inplace=True)
df.reset_index()
```

```python
df.fillna(0, inplace=True)
df
```

```python
# some of the columns are float type and some are object type. 
# certain values are str type and can not be converted into float type easily
# let's convert everything into str and take care those special cases first.
# then convert them all into float type
```

```python
for col in range(1,len(df.columns)):
    df.iloc[:,col] = [str(x) for x in df.iloc[:,col]]
```

```python
# take care those special cases - '51.56.3'. split them and put back the first 2.
# and also take care values that contains not just digit and '.'
for col in range(1,len(df.columns)):
    for row in range(len(df)):
        df.iloc[row,col] = df.iloc[row,col].split('.')[0]+'.'+df.iloc[row,col].split('.')[1]
        df.iloc[row,col] = re.sub('[^\d.]','',df.iloc[row, col])
```

```python
df
```

```python
# convert all the values to float type
for col in range(1,len(df.columns)):
    df.iloc[:,col] = [float(x) for x in df.iloc[:,col]]
```
